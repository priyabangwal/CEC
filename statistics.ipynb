```python
sns.boxplot(data = df , x = df['hp'])         #
```

```python
df['hp'].describe()
```

```python
q1 = df['hp'].quantile(0.25)
q3 = df['hp'].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
iqr = q3 - q1
iqr
```

```python
u = q3 + 1.5 * iqr            #high outliers
l = q1 -1.5 * iqr             #low outliers
```

```python
u
```

```python
l
```

```python
filter = df['hp'] > u
df[filter]
```

```python
filt = df['hp'] < u
df2 = df[filt]
df2
```
